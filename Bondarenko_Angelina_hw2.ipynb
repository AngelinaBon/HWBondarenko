## 10. Много разных строк.

Напишите функцию` get_extended_string_info`, принимающую на вход строку, а на выход выдаёт **7 строк через запятую**, полученные из оригинальной соответствующим преобразованием:

1. Самое длинное слово в строке
2. Количество слов в строке
3. Переворачивает каждое слово в строке задом наперёд
4. Удаляет все цифры из строки
5. Заменяет все буквы верхнего регистра на нижний, а нижнего регистра на верхний
6. Проверяет, является ли строка палиндромом (без учета пробелов, регистра и пунктуации, те только буквы и цифры)
7. Если строка не является палиндромом, то выводит первые 5 символов строки в обратном порядке

Т.е.  первая строка - самое длинное слово в исходной строке, вторая строка - количество слов в исходной строке и.т.д.

```python
# YOUR CODE HERE
def get_extended_string_info(s: str):
  s_splited = s.split()
  longest_word = max(s_splited, key=len)

  number_words = len(s_splited)

  revers = ''
  for x in s_splited:
    revers = revers + x[::-1] + ' '

  no_numbers = ''.join (x for x in s if (x.isdigit() == False))

  registr = s.swapcase()

  #проверка на палиндром
  s_modified = s.replace(" ", "").lower()
  s_without = ''
  s_without = ''.join (x for x in s_modified if x.isalnum())
  if s_without == s_without[::-1]:
    if_palindrom = 'строка является палиндромом'
  else:
    if_palindrom = 'строка не является палиндромом'

  #Если строка не является палиндромом, то выводит первые 5 символов строки в обратном порядке
  if if_palindrom == 'строка является палиндромом':
    symbol5 = 'для такой строки нет значения по условию, тк она является палиндромом'
  else:
    symbol5 = s[:5][::-1]
    #Этот вариант для случая, когда мы считаем, что пробел - тоже символ.
    #Если в задании подразумевалось обратное, то строку выше можно модифицировать, удалив из нее пробелы через s.replace(" ", "") и взять тот же срез [:5][::-1]

  return (longest_word, str(number_words), revers, no_numbers, registr, if_palindrom, symbol5)
```
